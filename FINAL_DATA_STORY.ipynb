                                            ''ΟΛΑ ΤΟΥ ΓΑΜΟΥ ΔΥΣΚΟΛΑ''


Από την αρχαιότητα μέχρι και τους πιο πρόσφατους καιρούς, ο γάμος είχε μία ξεχωριστή σημασία στις τάξεις της κοινωνίας. Αποτελούσε την επισφράγηση και τη νομιμοποίηση - αν προτιμάτε - του έρωτα μεταξύ δύο ανθρώπων. Με την πάροδο των χρόνων, ο θεσμός του γάμου άρχισε να φθείρεται. Σήμερα, τα ζευγάρια επιλέγουν να συζούν, να συγκατοικούν, αλλά αποφεύγουν να παντρεύονται είτε με θρησκευτικό είτε με πολιτικό γάμο. Το φαινόμενο, ωστόσο, δεν παρατηρείται μόνο στην Ελλάδα αλλά και στην υπόλοιπη Ευρώπη. Η τελευταία δεκαετία, που αποτελεί και το αντικείμενο της μελέτης μας, χαρακτηρίζεται από την περίοδο της κρίσης, της οικονομικής ύφεσης και των ανατροπών που έφερε στις ζωές μας η πανδημία της Covid-19. Ποιος ήταν ο ρόλος τους στη διαμόρφωση αυτής της κατάστασης; 

Ειδικότερα στην Ελλάδα, όπως μας εξηγεί ο καθηγητής Δημογραφίας του Πανεπιστημίου της Θεσσαλίας, Γιώργος Κοντογιάννης, το φαινόμενο της μείωσης των γάμων έχει τις ρίζες του στη δεκαετία του 1980. Σημαντικοί παράγοντες αυτής της μεταβολής ήταν οι εκσυγχρονιστικές μεταρρυθμίσεις της τότε κυβέρνησης. Για παράδειγμα, η θεσμοθέτηση του πολιτικού γάμου (νόμος 1329), η νομιμοποίηση των αμβλώσεων (1986) και η πρόβλεψη για ίση αμοιβή και ίση εργασία γυναικών και αντρών (1984) αποτελούν τομείς στα ζητήματα ισότητας και χειραφέτησης των γυναικών. "Η αλλαγή στο τρόπο τεκνοποίησης έπαιξε καθοριστικό ρόλο στη διαμόρφωση των σημερινών στατιστικών στοιχειών. Τα νέα ζευγάρια καθυστερούν την τεκνοποίηση ή επιλέγουν να αποκτήσουν παιδιά εκτός γάμου", επισημαίνει επιπρόσθετα ο κ. Κοντογιάννης. Το 2010 ο αριθμός των γάμων στη χώρα μας ανέρχεται στις 56.338 για να φτάσουμε στις 31.475 το 2020. 




Στη μείωση των γάμων σημαντικό ρόλο συντέλεσε και η νομοθετική ρύθμιση που αφορούσε τα σύμφωνα συμβίωσης (νόμος 4356/2015). Μέχρι τότε, η σύναψη συμφώνων συμβίωσης στη χώρα μας δεν είχε γνωρίσει ιδιαίτερη ανάπτυξη. Χαρακτηριστικά αναφέρουμε πως το 2011 υπολογίζονται μόλις 185 σύμφωνα συμβίωσης, με τον αριθμό να εκτοξεύεται στα 8.986 το 2020. Φυσικά στο σημείο αυτό πρέπει να τονισθεί πως η σύναψη συμφώνων συμβίωσης επιτρέπεται και στα ομόφυλλα ζευγάρια, κάτι που δεν συμβαίνει με τους γάμους. Παρά ταύτα, ο κ. Κοντογιάννης επισημαίνει πως αρκετά ετερόφυλλα ζευγάρια επιλέγουν την συγκεκριμένη μέθοδο αντί να προχωρήσουν στη τέλεση γάμου. Αυτό οφείλεται κυρίως σε οικονομικούς λόγους. 



Παρόμοιες τάσεις ανινχεύονται και σε ευρωπαϊκές χώρες. Συγκεκριμένα, στη παρούσα μελέτη, εξετάστηκαν χώρες κράτη - μέλη της Ευρωπαϊκής Ένωσης με πληθυσμό αντίστοιχο της Ελλάδας, ώστε τα αποτελέσματα να είναι συγκρίσημα. Χώρες όπως η Τσεχία (10.494.836), η Πορτογαλία (10.298.252) και η Σουηδία (10.379.295) εμφανίζουν την ίδια πτωτική πορεία. Εξαίρεση αποτελεί η Ουγγαρία (9.730.772), όπου από το 2012 και μετά υπάρχει θεαματική αύξηση στους γάμους με κορύφωση το 2020, τη χρονιά δηλαδή του lockdown, όπου τα υπόλοιπα κράτη παρουσιάζουν και το χαμηλοτερό αριθμό γάμων. 




Η Ουγγαρία, βέβαια, είναι μία ιδιάζουσα περίπτωση. Η κυβέρνηση του Βίκτορος Όρμπαν στη προσπάθεια της να αντιμετωπίσει το φαινόμενο της υπογεννητικότητας, που αποτελεί μάστιγα για την χώρα, το φθινόπωρο του 2019 νομοθέτησε κρατική επιχορήγηση ύψους 30.000 ευρώ σε νεόνυμφα ζευγάρια κάτω των σαράντα ενός ετών. Σκοπός αυτού του μέτρου είναι να αποτελέσει κίνητρο για τα ζευγάρια να αποκτήσουν παιδί, αφού το ποσό δίνεται με τη μορφή δανείου. Αν το ζευγάρι καταφέρει στα πέντε πρώτα χρόνια να κάνει παιδί, ο τόκος μηδενίζεται και αν αυτό συμβεί στην τριετία, το δάνειο τους χαρίζεται. Πράγματι, όπως αποδεικνύεται και από το σχετικό γράφημα, η Ουγγαρία πέτυχε το σκοπό της με τους γάμους να σημειώνουν ρεκόρ σαρανταετίας.

Ποια είναι όμως η ενδιάμεση ηλικία όσων παντρεύονται; Στα στοιχεία που καταγράφει η EUROSTAT και αφορούν την ενδιάμεση ηλικία των συζύγων τη στιγμή του γάμου, η διαφορά ηλικίας του ζευγαριού είναι περίπου στα 2,5 - 3,5 χρόνια. Τα νέα ζευγάρια επιλέγουν να νυμφευθούν σε μεγαλύτερη ηλικία σε σχέση με το παρελθόν. Αυτό οφείλεται τόσο στις οικονομικές - κοινωνικές μεταβολές στη διάρκεια των χρόνων όσο και στο γεγονός ότι ο γάμος έπαψε να έχει τον θεσμικό χαρακτήρα του παρελθόντος. Στη χώρα μας, το 2011, η ενδιάμεση ηλικία των παντρεμένων αντρών ήταν τα 32,7 έτη και των γυναικών τα 29,4 έτη. Δέκα χρόνια μετά, η ηλικία του άντρα παρουσιάζει μικρή μεταβολή κατά ένα χρόνο κι ένα μήνα και της γυναίκας κατά ένα χρόνο και τρεις μήνες. Η Πορτογαλία είναι η χώρα με την μεγαλύτερη μεταβολή στην ενδιάμεση ηλικία των συζύγων, αφού εμφανίζει διαφοροποίηση τεσσάρων ετών τόσο στους άντρες όσο και στις γυναίκες (2011: 30,1 και 28 αντίστοιχα, 2020: 34 και 32.8 αντίστοιχα).




Η ευθραυστότητα του γάμου, βέβαια, δεν αποδεικνύεται μόνο από τη μείωση των γαμήλειων τελετών. Ιδιαίτερη εντύπωση προκαλεί και ο αυξανόμενος αριθμός των διαζυγίων. Καθημερινά, όλο και περισσότερα ζευγάρια οδηγούνται στον χωρισμό. Οι κοινωνικές αλλαγές είχαν αντίκτυπο στη διατήρηση των γάμων. Ο ρόλος της γυναίκας έχει ενισχυθεί και οι απαιτήσεις του ζευγαριού μέσα στο γάμο έχουν αλλάξει.Την ίδια στιγμή το διαζύγιο δεν συνδέεται με το στίγμα που ενδεχομένως άφηνε στο παρελθόν. Άλλωστε και η νομοθεσία κατά τη διάρκεια των χρόνων παρουσίασε μεταβολές, προκειμένου τα ζευγάρια που επιλέγουν να λήξουν το γάμο τους συναινετικά, να μπορούν να το κάνουν μέσα από μία απλουστευμένη διαδικασία. Χαρακτηριστικό παράδειγμα η από κοινού υπουργική απόφαση των υπουργών επικρατείας και ψηφιακής διακυβέρνησης και δικαιοσύνης (ΦΕΚ Β'6390/31.12.2021) επιτρέπει την έκδοση συναινετικού διαζυγίου μέσω της πλατφόρμας gov.gr.

Στην Ελλάδα τα διαθέσιμα στοιχεία της ΕΛΣΤΑΤ για τα διαζύγια σταματούν στο 2017, ενώ για τις υπόλοιπες ευρωπαϊκές χώρες υπάρχουν δεδομένα μέχρι το 2020. Ο καθηγητής δημογραφίας του Πανεπιστημίου Θεσσαλίας μας ανέφερε πως περίπου 2,5 - 3 στους δέκα γάμους στη χώρα μας οδηγούνται σε διαζύγιο τη δεκαετία 2010-2020. Ενδεικτικά μόνο το 2017 εκδόθηκαν 8.177 περισσότερα διαζύγια σε σχέση με το προηγούμενο έτος. Όσον αφορά την Ευρώπη, εκτός από την Σουηδία, οι υπόλοιπες τρεις χώρες μελέτης παρουσιάζουν μείωση στον ετήσιο αριθμό διαζυγίων το 2020, γεγονός που σε μεγάλο βαθμό οφείλεται και στην υγειονομική κρίση.




Μιλώντας για το μέλλον και τις προκλήσεις που τα στατιστικά στοιχεία φέρνουν, ο κ. Κοντογιάννης σημειώνει:"τα επόμενα χρόνια η διαφοροποίηση της δομής της οικογένειας θα γίνει περισσότερο εμφανής. Οι μονογονεϊκές οικογένειες θα αυξηθούν περαιτέρω. Το ίδιο θα συμβεί και με τις οικογένειες που θα έχουν ένα θετό κι έναν βιολογικό γονέα". Σημειώνει δε ότι η αποσύνδεση του παιδιού από το γάμο που επιτρέπει στα ζευγάρια να δημιουργούν οικογένεια χωρίς να χρειάζεται να παντρευτούν, θα επιτρέψει την συνέχιση της πτωτικής πορείας των γάμων στην Ελλάδα. Η μείωση παιδιών και η αύξηση των διαζυγίων αναπόφευκτα σε μερικά χρόνια από σήμερα θα οδηγήσει σε ανθρώπους που θα μεγαλώνουν ηλικιακά μόνοι και σε μικρό οικογενειακό περιβάλλον, δηλαδή χωρίς αδέλφια, ξαδέλφια, ανήψια κτλ.

Αναμφισβήτητα η οικονομική κρίση και η πανδημία έπαιξαν τον ρόλο τους στη διαμόρφωση των σημερινών δεδομένων. Οι κοινωνικές μεταβολές που σημειώθηκαν από τη δεκαετία του 80' και μετέπειτα απελευθέρωσαν τη γυναίκα από το ρόλο της συζύγου και της μητέρας αποκλειστικά εντός του γάμου. Επιπλέον, ο γάμος έπαψε να είναι υποχρέωση αλλά συναίσθημα, όπως γράφει στο βιβλίο του "Ο γάμος από έρωτα έχει αποτύχει;" ο Πασκάλ Μπρυκνέρ. Κι όμως αυτή η "επανάσταση" που συντελέστηκε, σύμφωνα με τον Μπρυκνέρ, ιερός να θεωρείται ο έρωτας και όχι ο γάμος, είχε τα αντίθετα από τα αναμενόμενα αποτελέσματα. Κατά τον Γάλλο συγγραφέα αυτό οφείλεται στις δυσάρεστες περιπέτειες του γάμου από έρωτα που επιβεβαιώνουν τη δυσκολία των ανθρώπων να συμβιώνουν.
