Η εργασία μας αφορά την μείωση των γάμων στη χώρα μας, γεγονός που προκύπτει από τα στατιστικά στοιχεία της ΕΛΣΤΑΤ, τα οποία
είναι διαθέσιμα στον ιστότοπό της. Η έρευνα επικεντρώθηκε στη τελευταία δεκαετία για δύο βασικούς λόγους:
α)τα δεδομένα της ΕΛΣΤΑΤ αφορούσαν τη δεδομένη χρονική περίοδο
β)θέλαμε να έχουμε μια επαρκή περίοδο για τον εντοπισμό των διαφοροποιήσεων και την εξαγωγή συμπερασμάτων.

Πηγές και συλλογή δεδομένων:
Βασική πηγή των δεδομένων που συλλέξαμε ήταν η ΕΛΣΤΑΤ. Από εκεί και πέρα, και λόγω του ενδιαφέροντός μας να διευρύνουμε την
έρευνά μας σε χώρες της Ευρώπης, ανατρέξαμε στα στατιστικά στοιχεία της EUROSTAT. Επομένως, συλλέξαμε του πληθυσμούς των 
ευρωπαϊκών κρατών, ώστε να βρούμε αυτές με πληθυσμό αντίστοιχο της χώρας μας. Εν συνεχεία, ψάξαμε τα δεδομένα για τους γάμους
στις χώρες αυτές, τις ηλικίες αντρών και γυναικών τη στιγμή του γάμου, καθώς και τον αριθμό διαζυγίων. Από τη σελίδα της ΕΛΣΤΑΤ,
επίσης, συλλέξαμε τον αριθμό των συμφώνων συμβίωσης.

Καθαρισμός δεδομένων - Ανάλυση δεδομένων:
Ο αριθμός των γάμων των υπο-εξέταση χωρών,δεν δινόταν σε απόλυτους αριθμούς, αλλά ανά χιλίους κατοίκους. Συνεπώς, και για την
διευκόλυνση της εργασίας μας, πολλαπλασιάσαμε τον πληθυσμό με τον αριθμό των γάμων ανά χίλιους κατοίκους και το διαιρέσαμε
με το χίλια.
Για να βρούμε τα κράτη που μας ενδιέφεραν, επιλέξαμε αυτά με πληθυσμό από εννιά έως έντεκα εκατομμύρια.
Για τα σύμφωνα συμβίωσης, επεξεργαστήκαμε το pdf με τα στοιχεία στο tabula και κρατήσαμε όσα μας ενδιέφεραν για την εκπόνιση
της μελέτης μας. 
Η ενδιάμεση ηλικία των συζύγων, όπως και ο αριθμός των διαζυγίων ανά έτος υπήρχαν στην σελίδα της EUROSTAT έτοιμα προς ανάλυση.
Τα αρχεία excel που κατεβάσαμε τα μετατρέψαμε σε αρχεία csv προκειμένου να τα εισάγουμε σε DataFrame στα pandas.
