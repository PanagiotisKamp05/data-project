                                             "ΟΛΑ ΤΟΥ ΓΑΜΟΥ ΔΥΣΚΟΛΑ"
                                    
                                                                                      

Από την αρχαιότητα μέχρι και τους πιο πρόσφατους καιρούς, ο γάμος είχε μία ξεχωριστή σημασία στις τάξεις της κοινωνίας. Αποτελούσε την επσφράγηση και τη νομιμοποίηση - αν προτιμάτε - του έρωτα μεταξύ δύο ανθρώπων. Με την πάροδο των χρόνων, ο θεσμός του γάμου άρχισε να φθείρεται. Σήμερα, τα ζευγάρια επιλέγουν να συζούν, να συγκατοικούν, αλλά στο άκουσμα του γάμου... "στρίβειν δια του αρραβώνος". Φόβος; Αγωνία; Αδιαφορία; Ή μήπως πράγματι ο γάμος σκοτώνει τον έρωτα; 

Την τελευταία δεκαετία στην Ελλάδα παρατηρείται σημαντική μείωση των γάμων. Πρόκειται για την περίοδο της κρίσης, της οικονομικής ύφεσης, των μνημονίων αλλά και των ανατροπών που έφερε στις ζωές μας η πανδημία της Covid-19. Ωστόσο, όπως μας επισημαίνει και ο καθηγητής Δημογραφίας του Πανεπιστημίου της Θεσσαλίας, Γιώργος Κοντογιάννης, το φαινόμενο της μείωσης των γάμων έχει τις ρίζες του στη δεκαετία του 1980. "Η αλλαγή στο τρόπο τεκνοποίησης έπαιξε καθοριστικό ρόλο στη διαμόρφωση των σημερινών στατιστικών στοιχειών. Τα νέα ζευγάρια καθυστερούν την τεκνοποίηση ή επιλέγουν να αποκτήσουν παιδιά εκτός γάμου". Ταυτόχρονα, η πολιτεία έδωσε τη δυνατότητα της επισύναψης συμφώνων συμβίωσης, που χρόνο με το χρόνο αποκτούν μεγαλύτερη αναγνωρισμότητα. "Τα σύμφωνα συμβίωσης, σήμερα, αποτελούν μία μορφή επισημοποίησης της σχέσης ενός ζευγαριού και αντικαθιστά τον θρησκευτικό γάμο, ακόμα και τον πολιτικό", σημειώνει ο κ. Κοντογιάννης. Οι μεταβολές που σημειώνονται στους αριθμούς, όπως παρατηρούμε στους δύο πίνακες παρακάτω, είναι χαρακτηριστικές. Το 2010 ο αριθμός των γάμων στη χώρα μας ανέρχεται στους 56.338 για να φτάσουμε στους 31.475 το 2020, σε αντίθεση με τα σύμφωνα συμβίωσης που το 2011 υπολογίζονται στα 185, ενώ το 2020 αγγίζουν τις 8.986.         

Βέβαια, οι τάσεις αυτές δεν ανιχνεύονται μόνο στην ελληνική κοινωνία. Παρόμοιες είναι οι καταστάσεις και σε χώρες της Ευρωπαϊκής Ένωσης με πληθυσμό αντίστοιχο της Ελλάδας. Συγκεκριμένα, χώρες όπως η Τσεχία (10.494.836), η Πορτογαλία (10.298.252) και η Σουηδία (10.379.295) εμφανίζουν την ίδια πτωτική πορεία. Μοναδική εξαίρεση αποτελεί η Ουγγαρία (9.730.772), όπου από το 2012 και μετά υπάρχει θεαματική αύξηση στους γάμους με κορύφωση το 2020, την χρονιά δηλαδή του lockdown όπου η Ουγγαρία απαρριθμεί 67.409 γάμους έναντι 44.914 της Τσεχίας, 18.532 της Πορτογαλίας, 37.179 της Σουηδίας και 31.475 της Ελλάδας. 





Σύμφωνα με τον κ. Κοντογιάννη, στην έρευνα που διενήργησε ο ίδιος για λογαριασμό του ΕΛ.ΙΔ.Ε.Κ (Ελληνικό Ίδρυμα Έρευνας και Καινοτομίας) διαπιστώνεται ότι στη πλειοψηφία οι γυναίκες είναι εκείνες που αναβάλλουν συχνότερα τους γάμους ή αρνούνται να παντρευτούν. Στα στοιχεία που καταγράφει η EUROSTAT και αφορούν την ενδιάμεση ηλικία των συζύγων τη στιγμή του γάμου, η διαφορά ηλικίας του ζευγαριού είναι περίπου στα 2,5 - 3,5 χρόνια. Απόρροια των όλων προηγούμενων είναι ότι τα ζευγάρια επιλέγουν να παντρευτούν σε μεγαλύτερη πλέον ηλικία συγκριτικά με το παρελθόν. Στη χώρα μας, το 2011, η ενδιάμεση ηλικία των νημφευμένων αντρών ήταν τα 32,7 έτη και των γυναικών τα 29,4 έτη. Δέκα χρόνια μετά, η ηλικία του άντρα παρουσιάζει μικρή μεταβολή κατά ένα χρόνο κι ένα μήνα και της γυναίκας κατά ένα χρόνο και τρεις μήνες. Η Πορτογαλία είναι η χώρα με την μεγαλύτερη μεταβολή στην ενδιάμεση ηλικία των συζύγων, αφού εμφανίζει διαφοροποίηση τεσσάρων ετών τόσο στους άντρες όσο και στις γυναίκες (2011: 30,1 και 28 αντίστοιχα, 2020: 34 και 32.8 αντίστοιχα). 





Η ευθραυστότητα του γάμου, βέβαια, δεν αποδεικνύεται μόνο από τη μείωση των γαμήλειων τελετών. Ιδιαίτερη εντύπωση προκαλεί και ο αυξανόμενος αριθμός των διαζυγίων. Καθημερινά, όλο και περισσότερα ζευγάρια οδηγούνται στον χωρισμό. Οι κοινωνικές αλλαγές είχαν αντίκτυπο στη διατήρηση των γάμων. Ο ρόλος της γυναίκας έχει ενισχυθεί και οι απαιτήσεις του ζευγαριού μέσα στο γάμο έχουν αλλάξει.Την ίδια στιγμή το διαζύγιο δεν συνδέεται με το στίγμα που ενδεχομένως άφηνε στο παρελθόν. Άλλωστε και η νομοθεσία κατά τη διάρκεια των χρόνων παρουσίασε μεταβολές, προκειμένου τα ζευγάρια που επιλέγουν να λήξουν το γάμο τους συναινετικά, να μπορούν να το κάνουν μέσα από μία απλουστευμένη διαδικασία. Χαρακτηριστικό παράδειγμα η από κοινού υπουργική απόφαση των υπουργών επικρατείας και ψηφιακής διακυβέρνησης και δικαιοσύνης (ΦΕΚ Β'6390/31.12.2021) επιτρέπει την έκδοση συναινετικού διαζυγίου μέσω της πλατφόρμας gov.gr. 

Στην Ελλάδα τα διαθέσιμα στοιχεία της ΕΛΣΤΑΤ για τα διαζύγια σταματάν στο 2017, ενώ για τις υπόλοιπες ευρωπαϊκές χώρες υπάρχουν δεδομένα μέχρι το 2020. Ο καθηγητής δημογραφίας του Πανεπιστημίου Θεσσαλίας μας ανέφερε πως περίπου 2,5 - 3 στους δέκα γάμους στη χώρα μας οδηγούνται σε διαζύγιο τη δεκαετία 2010-2020. Ενδεικτικά μόνο το 2017 εκδόθηκαν 8.177 περισσότερα διαζύγια σε σχέση με το προηγούμενο έτος. Όσον αφορά την Ευρώπη, εκτός από την Σουηδία, οι υπόλοιπες τρεις χώρες μελέτης παρουσιάζουν μείωση στον ετήσιο αριθμό διαζυγίων το 2020, γεγονός που σε μεγάλο βαθμό οφείλεται και στην υγειονομική κρίση. 







Μιλώντας για το μέλλον και τις προκλήσεις που τα στατιστικά στοιχεία φέρνουν, ο κ. Κοντογιάννης σημειώνει:"τα επόμενα χρόνια η διαφοροποίηση της δομής της οικογένειας θα γίνει περισσότερο εμφανής. Οι μονογονεϊκές οικογένειες θα αυξηθούν περαιτέρω. Το ίδιο θα συμβεί και με τις οικογένειες που θα έχουν ένα θετό κι έναν βιολογικό γονέα". Σημειώνει δε ότι η αποσύνδεση του παιδιού από το γάμο που επιτρέπει στα ζευγάρια να δημιουργούν οικογένεια χωρίς να χρειάζεται να παντρευτούν, θα επιτρέψει την συνέχιση της πτωτικής πορείας των γάμων στην Ελλάδα. Η μείωση παιδιών και η αύξηση των διαζυγίων αναπόφευκτα σε μερικά χρόνια από σήμερα θα οδηγήσει σε ανθρώπους που θα μεγαλώνουν ηλικιακά μόνοι και σε μικρό οικογενειακό περιβάλλον, δηλαδή χωρίς αδέλφια, ξαδέλφια, ανήψια κτλ. 

Αναμφισβήτητα η οικονομική κρίση και η πανδημία έπαιξαν τον ρόλο τους στη διαμόρφωση των σημερινών δεδομένων. Οι κοινωνικές μεταβολές που σημειώθηκαν από τη δεκαετία του 80' και μετέπειτα απελευθέρωσαν τη γυναίκα από το ρόλο της συζύγου και της μητέρας αποκλειστικά εντός του γάμου. Επιπλέον, ο γάμος έπαψε να είναι υποχρέωση αλλά συναίσθημα, όπως γράφει στο βιβλίο του "Ο γάμος από έρωτα έχει αποτύχει;" ο Πασκάλ Μπρυκνέρ. Κι όμως αυτή η "επανάσταση" που συντελέστηκε, σύμφωνα με τον Μπρυκνέρ, ιερός να θεωρείται ο έρωτας και όχι ο γάμος, είχε τα αντίθετα από τα αναμενόμενα αποτελέσματα. Κατά τον Γάλλο συγγραφέα αυτό οφείλεται στις δυσάρεστες περιπέτειες του γάμου από έρωτα που επιβεβαιώνουν τη δυσκολία των ανθρώπων να συμβιώνουν.
